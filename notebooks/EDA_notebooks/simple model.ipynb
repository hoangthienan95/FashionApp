```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable
from torch import nn

class DenseNet(nn.Module):

    def __init__(self, n_items, n_factors, H1, D_out):
        """
        Simple Feedforward with Embeddings
        """
        super().__init__()
   	# item embedding layers
        self.item_factors = torch.nn.Embedding(n_items, n_factors,
                                               sparse=True)
   	# linear layers
        self.linear1 = torch.nn.Linear(n_factors*2, H1)
        self.linear2 = torch.nn.Linear(H1, D_out)

    def forward(self, items):
        items_embedding = self.item_factors(items)

        h1_relu = F.relu(self.linear1(items_embedding))
        output_scores = self.linear2(h1_relu)
        return output_scores

    def predict(self, items):
        # return the score
        output_scores = self.forward(items)
        return output_scores
```
